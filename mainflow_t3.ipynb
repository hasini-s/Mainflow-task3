```python
import pandas as pd
#reading the database
data=pd.read_csv("tips.csv")
#printing the top 10 rows
data.head(10)
```

```python
import matplotlib.pyplot as plt
#scatter plot with day against tip
plt.scatter(data['day'],data['tip'])
#adding title to the plot
plt.title('Scatter Plot')
#Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
#Create the scatter plot with day on the x-axis and tip on the y-axis plt.scatter(data['day'], data['tip'])
#Add a colorbar to the plot based on the 'size' column
plt.colorbar(plt.scatter(data['day'], data['tip'], c=data['size']))
# Add a title and labels to the plot
plt.title('Scatter Plot with Colorbar')
plt.xlabel('Day')
plt.ylabel('Tip')
# Show the plot
plt.show()
```

```python
# Scatter plot with day against
plt.plot (data['tip'])
plt.plot (data['size'])
# Adding Title to the Plot
plt.title(" Scatter Plot")
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
# Bar chart with day against tips
plt.bar (data['day'], data ['tip'])
plt.title("Bar Chart")
# Setting the X and Y labels
plt.xlabel ('Day')
plt.ylabel( 'Tip')
plt.show()
```

```python
plt.hist(data['tip'])
plt.title('Histogram of Tips')
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.show()
```
